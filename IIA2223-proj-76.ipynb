### Classe JogoBT_76(Game)
Decidi que quando uma peça é capturada a mesma é eliminada do dicionario correspondente 

````python
##########class jogo_BT(Game):
class JogoBT_76(Game):
    """Play breakthrough on an 8 x 8 board"""

    def __init__(self):
        self.fullboard = set([(x, y) for x in range(1, 9)
                 for y in range(1, 9)])
        self.initial = EstadoBT_76(to_move = 1, white = {'1': ('a',1),
    '2': ('b',1),
    '3': ('c',1),
    '4': ('d',1),
    '5': ('e',1),
    '6': ('f',1),
    '7': ('g',1),
    '8': ('h',1),
    '9': ('a',2),
    '10': ('b',2),
    '11': ('c',2),
    '12': ('d',2),
    '13': ('e',2),
    '14': ('f',2),
    '15': ('g',2),
    '16': ('h',2)}, blacks={'1': ('a',8),
    '2': ('b',8),
    '3': ('c',8),
    '4': ('d',8),
    '5': ('e',8),
    '6': ('f',8),
    '7': ('g',8),
    '8': ('h',8),
    '9': ('a',7),
    '10': ('b',7),
    '11': ('c',7),
    '12': ('d',7),
    '13': ('e',7),
    '14': ('f',7),
    '15': ('g',7),
    '16': ('h',7)})
    
    def actions(self, state):
        "Legal moves"
        return state.moves()

    ## Player who´s turn is it
    def to_move(self, state):
        return state.to_move

    def result(self, state, move):
        whiteCopy = state.white.copy()
        blackCopy = state.blacks.copy()
        delete = []


        if state.to_move == 1:
            for w in whiteCopy:
                if (move[0],int(move[1])) == whiteCopy[w]:
                    whiteCopy[w] = (move[3], int(move[4]))
            for c in blackCopy:
                if blackCopy[c] == (move[3], int(move[4])):
                    delete.append(c)
            for i in delete:
                del blackCopy[i]


                    
        else:
            for b in blackCopy:
                if (move[0],int(move[1])) == blackCopy[b]:
                    blackCopy[b] = (move[3], int(move[4]))
            for d in whiteCopy:
                if whiteCopy[d] == (move[3], int(move[4])):
                    delete.append(d)
            for i in delete:
                del whiteCopy[i]


        return EstadoBT_76(to_move=(2 if state.to_move == 1 else 1),
                         white=whiteCopy,blacks=blackCopy) 

    def utility(self, state, player):
        aux = state.compute_utility(player)
        return True if player == aux else False

    def terminal_test(self, state):
        return state.terminou != 0
    
    def executa(self, estado, listaJogadas):
        estadoTemp = estado
        for m in listaJogadas:
            if m in self.actions(estadoTemp) and not self.terminal_test(estadoTemp):
                estadoTemp = self.result(estadoTemp, m)
                self.initial = estadoTemp
                self.to_move = self.initial.to_move
        return estadoTemp


    def display(self, state):
        state.display()
        fim = self.terminal_test(state)
        if  fim:
            pass
        else :
            print("--NEXT PLAYER: ", end='')
            if state.to_move == 1:
                print('W')
            else:
                print('B')
    

    
````

A heuristica que escolhi Ronaldo/func_aval_76 consiste em:

1. Contagem de peças: Um lado ganha pontos extras se tiver mais peças no tabuleiro. 10 pontos para cada peça extra. 

2. Avalia a posição das peças: Durante este processo, cada peça será avaliada de acordo com sua localização. Além disso, o valor deve ser menor nas bordas. 
3. Recompensar as peças seguras: As peças seguras (não estão em perigo de serem capturadas por uma peça inimiga) teram um bónus de 50% com base no valor da sua posição.

````python

## primeira heuristica complexa que pensei
## mas esta demasiado confusa e rebenta com d= 3 e nsec = 10
def BoardEvaluation(state,player):
    terminado = 500000
    PieceHomeGroundValue = 10
    value = 0
    PieceDangerValue = 10
    PieceHighDangerValue = 100
    PieceAttackValue = 50

    if state.terminou == 1:
        return terminado if player == 1 else -terminado
    elif state.terminou == -1:
        return terminado if player == 2 else -terminado
        
    elif player == 1:
        for w in state.white:
            pieceValue = 1300
            ## evaluate attack
            if ((chr(ord(state.white[w][0])-1),state.white[w][1]+1) in state.blacks.values()) or ((chr(ord(state.white[w][0])+1),state.white[w][1]+1) in state.blacks.values()):
                pieceValue -= PieceAttackValue

            else:
                if state.white[w][1] == 5:
                    pieceValue += PieceDangerValue
                elif state.white[w][1] == 6:
                    pieceValue += PieceHighDangerValue
                elif state.white[w][1] == 2:
                    pieceValue += PieceDangerValue
                elif state.white[w][1] == 1:
                    pieceValue += PieceHighDangerValue
                else:
                    pieceValue += state.white[w][1] * PieceDangerValue
            if (state.white[w][1] == 1):
                pieceValue += PieceHomeGroundValue
            value += pieceValue

    else:
        for b in state.blacks:
            pieceValue = 1300
            if ((chr(ord(state.blacks[b][0])-1),state.blacks[b][1]-1) in state.white.values()) or ((chr(ord(state.blacks[b][0])+1),state.blacks[b][1]-1) in state.white.values()):
                pieceValue -= PieceAttackValue

            else:
                if state.blacks[b][1] == 4:
                    pieceValue += PieceDangerValue
                elif state.blacks[b][1] == 3:
                    pieceValue += PieceHighDangerValue
                elif state.blacks[b][1] == 7:
                    pieceValue += PieceDangerValue
                elif state.blacks[b][1] == 8:
                    pieceValue += PieceHighDangerValue
                else:
                    pieceValue += (9-state.blacks[b][1]) * PieceDangerValue
            if (state.blacks[b][1] == 8):
                pieceValue += PieceHomeGroundValue
            value += pieceValue
            
    value += len(state.moves())
    return value
````

````python
from random import *

## Heuristica defensiva que valoriza o maior numero de pecas
def defensivo(state,player):

    terminado  = 10000
    if state.terminou == 1:
        return terminado if player == 1 else -terminado
    elif state.terminou == -1:
        return terminado if player == 2 else -terminado

    if player == 1:
        return 2 * len(state.white) + random()
    else:
        return 2 * len(state.blacks) + random()
        
## Heuristica ofensiva que valoriza o ataque a pecas adversarias
def ofensivo(state,player):

    terminado  = 10000


    if state.terminou == 1:
        return terminado if player == 1 else -terminado
    elif state.terminou == -1:
        return terminado if player == 2 else -terminado

    if player == 1:
        return 2 * (30-len(state.blacks)) + random()
    else:
        return 2 * (30-len(state.white)) + random()


  
````

````python
from random import *    

def belarmino(state,player):
    value = 0
    terminado  = 10000

    if state.terminou == 1:
        return terminado if player == 1 else -terminado
    elif state.terminou == -1:
        return terminado if player == 2 else -terminado

    if player == 1:
        for j in range(97,105):
            countP = 0
            for i in range(1,9):    
                if(chr(j),i) in state.white.values():
                    countP += 1
            value += pow(j-96,j-96) * countP
    else:
        for j in range(97,105):
            countP = 0
            for i in range(1,9):    
                if(chr(j),i) in state.blacks.values():
                    countP += 1
            value += pow(9-(j-96),9-(j-96)) * countP


    return value

    
````
